### Synopsis  

```python
okt = Okt()
reg = re.compile("[^---\s]")
tokenized_synopsis = []
with open("C:/datasets_temp/synopsis.txt", 'r', encoding='utf-8') as file:
    while True:
        line = file.readline()
        if not line: break
        
        # 
        line = reg.sub('', line)
        
        #  
        words = okt.pos(line, stem=True, norm=True)
        res = []
        for w in words:
            if w[1] in ["Noun"]:      # "Adjective", "Verb"    .
                res.append(w[0])    #   tokenized   . 
        tokenized_synopsis.append(res)
```

#### WEAT-score      

```python
def cos_sim(i, j):
    return np.dot(i, j.T)/(np.linalg.norm(i)*np.linalg.norm(j))
```

```python
def s(w, A, B):
    c_a = cos_sim(w, A)
    c_b = cos_sim(w, B)
    mean_A = np.mean(c_a, axis=-1)
    mean_B = np.mean(c_b, axis=-1)
    return mean_A - mean_B
```

```python
def weat_score(X, Y, A, B):
    
    s_X = s(X, A, B)
    s_Y = s(Y, A, B)

    mean_X = np.mean(s_X)
    mean_Y = np.mean(s_Y)
    
    std_dev = np.std(np.concatenate([s_X, s_Y], axis=0))
    
    return  (mean_X-mean_Y)/std_dev
```

```python
def get_vectors(words):
    return np.array([model.wv[word] for word in words if word in model.wv])
```

```python
#  
def plot_heatmap(matrix_df, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrix_df, annot=False, cmap="coolwarm", fmt=".3f", linewidths=0.5)
    plt.title(title)
    plt.xlabel("Class")
    plt.ylabel("Class")
    plt.xticks(rotation=45)
    plt.yticks(rotation=45)
    plt.show()
```

```python
# WEAT Score  (21, 21)  
n_classes = len(class_keywords.keys()) - 2  # synopsis_gen, synopsis_art 
genart_weat_matrix = np.zeros((n_classes, n_classes))

#     (Gen, Art )
filtered_classes = [c for c in class_keywords.keys() if c not in ["synopsis_gen.txt", "synopsis_art.txt"]]

vectors_gen = get_vectors(class_keywords["synopsis_gen.txt"])
vectors_art = get_vectors(class_keywords["synopsis_art.txt"])

#    (P, Q)  WEAT Score 
for i in range(n_classes):
    for j in range(n_classes):
        if i == j:
            genart_weat_matrix[i, j] = np.nan  #     X
        
        class_P = filtered_classes[i]
        class_Q = filtered_classes[j]

        vectors_P = get_vectors(class_keywords[class_P])
        vectors_Q = get_vectors(class_keywords[class_Q])

        # WEAT Score  (P vs Q - General vs Art)
        genart_weat_matrix[i, j] = weat_score(vectors_P, vectors_Q, vectors_gen, vectors_art)

genart_df = pd.DataFrame(
    genart_weat_matrix, 
    index=filtered_classes, 
    columns=filtered_classes
)
```

```python
# GEN-ART WEAT Score 
plot_heatmap(genart_df, "WEAT Score (Bias: General vs Art)")
```
